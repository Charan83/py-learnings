```python
dad_joke_url = "https://icanhazdadjoke.com/"

resp_html = requests.get(dad_joke_url) # by default it will send the html --> (default header is) headers={"Accept": "text/html"}
# print(resp_html.text)

resp_plain_text = requests.get(dad_joke_url, headers={"Accept": "text/plain"})
print(resp_plain_text.text)

resp_json = requests.get(dad_joke_url, headers={"Accept": "application/json"})
print(resp_json.text) # its a string "{}" and not an object
print(resp_json.json())# is a dictionary object
data = resp_json.json()
print(type(data))
print(data['joke'])
```

# Query String
- way to pass data to server as part of get request
- http://www.example.com/?key1=val1&key2=val2

```python
q_term = "cat"
dad_joke_search = dad_joke_url + "search"
res = requests.get(dad_joke_search, headers={"Accept": "application/json"}, params={'term': q_term, 'limit': 2})
data = res.json()
print(data)
```

```python
from pyfiglet import figlet_format
from termcolor import colored
import requests
import random

title = "Dad Joke 3000"
dad_joke_search_url = "https://icanhazdadjoke.com/search"
ascii_title = figlet_format(title)
ascii_colored_titile = colored(ascii_title, color='red')
print(ascii_colored_titile)

topic = input("Let me tell you a joke! Give me a topic : ")

def get_joke(topic):
    res = requests.get(dad_joke_search_url, 
                       headers={"Accept": "application/json"},
                       params={"term": topic}
                      )
    data = res.json()
    total_jokes = data['total_jokes']
    # print(data)
    msg = ""
    if total_jokes > 1:
        msg = f"I have got {total_jokes} jokes about {topic}. Here's one : "
        joke_num = random.randrange(len(data['results'])-1)
        joke = data['results'][joke_num]['joke']
        print(f"{msg}\n{joke}")
    elif total_jokes == 1:
        msg = f"I have got one joke about {topic}. Here's it is : "
        joke = data['results'][0]['joke']
        print(f"{msg}\n{joke}")
    else:
        msg = f"No jokes found for the topic : {topic}. Please try again!"
        
    
    
    
get_joke(topic)
```
